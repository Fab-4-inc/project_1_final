```python
crime_count = data["Type of Crime"].value_counts()

## Create a DATAFRAME ** 

crime_count = pd.DataFrame(crime_count)


#rename columns
crime_count = crime_count.rename(columns={"Type of Crime":"Crime Count"})
crime_count.index.names = ['Type of Crime']
crime_count

```

```python
#Neighborhood Crime
crime_by_neighborhood = data["Neighborhood"].value_counts()

## Create a DATAFRAME ** 

crime_by_neighborhood = pd.DataFrame(crime_by_neighborhood)
crime_by_neighborhood

#rename columns
crime_by_neighborhood = crime_by_neighborhood.rename(columns={"Neighborhood":"Crime Count"})
crime_by_neighborhood.index.names = ['Neighborhood']
crime_by_neighborhood.head()

```

```python
plt.figure(figsize=(33,5))
#bar chart: Crime Count in August 2019
crime_chart = crime_count.plot(kind='bar')

#X-Axis = Crime Type
crime_chart.set_xlabel("Type of Crime", fontsize=11)
crime_chart.set_ylabel("Crime Count", fontsize=11)

#y_pos = np.arange(len(bars))

#Title
crime_chart.set_title("Autugst 2019 Crime Count", fontsize=18)

#resizing
# Get current size
#fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
#print("Current size:", fig_size)
 
# Set figure width & height
##fig_size[0] = 20
#fig_size[1] = 5
#plt.rcParams["figure.figsize"] = fig_size

plt.legend(loc='upper right')

plt.xticks(ha='right', rotation=35, fontsize=7, fontname='monospace')


plt.savefig("August_2019_Crime_Count12",bbox_inches="tight")


plt.show()

#plt.tight_layout()

```
