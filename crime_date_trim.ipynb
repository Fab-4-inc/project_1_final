```python
# Dependencies
import pandas as pd
```

```python
data_file = "Crimes_-_Map.csv"
```

```python
# Use Pandas to read data
data_file_pd = pd.read_csv(data_file)
data_file_pd.head()
```

```python
data_file_pd.dtypes
```

```python
data_file_pd["DATE  OF OCCURRENCE"] = data_file_pd["DATE  OF OCCURRENCE"].astype('datetime64[ns]')
```

```python
data_file_pd.dtypes
```

```python
trim_crime = data_file_pd.loc[:,["DATE  OF OCCURRENCE", " PRIMARY DESCRIPTION", "LATITUDE","LONGITUDE"]]
trim_crime = trim_crime.dropna(how='any')
trim_crime
```

```python
trim_crime['DATE  OF OCCURRENCE'] = [time.date() for time in trim_crime['DATE  OF OCCURRENCE']]
```

```python
trim_crime
```

```python
year_month = pd.DatetimeIndex(trim_crime['DATE  OF OCCURRENCE']).to_period("M")
year_month
```

```python
trim_crime["Date (Year/Month)"] = year_month
trim_crime.head()
```

```python
project_crime_data = trim_crime[["Date (Year/Month)", " PRIMARY DESCRIPTION", "LATITUDE", "LONGITUDE"]]
project_crime_data.head()
```

```python
project_crime_data.to_csv("crime_project_data.csv")
```

```python
aug_2019_crime = project_crime_data.loc[project_crime_data["Date (Year/Month)"] == "2019-8"]
aug_2019_crime.head()
```

```python
aug_2019_crime.to_csv("aug_crime_2019.csv")
```
