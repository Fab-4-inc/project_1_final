```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
red_fin_data = "red_fin.csv"
```

```python
red_fin_data = "red_fin.csv"
house_df = pd.read_csv(red_fin_data)
house_df
```

```python
house_df["Year"] = house_df["Month of Period End"].str[-2:]
house_df
```

```python
summary_house = house_df[["Region","Year","Month of Period End","Median Sale Price", "Homes Sold", "New Listings","Days on Market"]]
summary_house
```

```python
summary_house = summary_house.rename(columns={"Month of Period End":"Month",
                                                  "Median Sale Price":"Sale Price"})
summary_house
```

```python
summary_house.dtypes
```

```python
summary_house['Region'] = summary_house['Region'].astype(str)
summary_house.dtypes
```

```python
houses_2019 = summary_house.loc[summary_house["Year"] == "19"]
houses_2019 = pd.DataFrame(houses_2019)
houses_2019['index'] = houses_2019.index
houses_2019.dropna()
houses_2019
```

```python
ch_neighborhoods =  (houses_2019["Region"].unique())
ch_neighborhoods
```

```python
len(ch_neighborhoods)
```

```python
houses_2019.head()
```

```python
neighbor_2019 = houses_2019.Region.str.split("-",expand=True)
neighbor_2019
```

```python
houses_2019["Neighborhood"] = neighbor_2019[1]
houses_2019
```

```python
houses_2019["City"] = neighbor_2019[0]
houses_2019
```

```python
summary_house_2 = houses_2019[["City","Neighborhood","Month","Sale Price", "Homes Sold", "New Listings","Days on Market"]]
summary_house_2
```

```python
aug_2019 = summary_house_2.loc[summary_house_2["Month"] == "Aug-19"]
aug_2019 = pd.DataFrame(aug_2019)
aug_2019.dropna()
aug_2019
```

```python
aug_2019.to_csv("Redfin_Aug_2019.csv",encoding="utf-8", index=False)
```
